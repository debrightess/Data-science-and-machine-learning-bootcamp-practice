### Notebook imports and headings

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Example 1: A simple cost function
$f(x) = x^2 + x + 1$

```python
def f(x):
  return x**2 + x + 1
```

```python
# Make data 
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
# Plot 
plt.xlim([-3, 3])
plt.ylim([0, 14])
plt.xlabel("x", fontsize=15)
plt.ylabel("f(x)", fontsize=15)
plt.plot(x_1, f(x_1))
plt.show()
```

## Slope and Derivatives 

```python
# Challenge: create a python function for the derivative of f(x) called df(x)
def df(x):
    return 2 * x + 1
```

```python
# Plot functions and derivatives side by side
plt.figure(figsize=[12,5])

# 1 Chart: Cost function
plt.subplot(1,2,1)
plt.xlim(-3, 3)
plt.ylim(0, 8)
plt.title("Cost function", fontsize=17)
plt.xlabel("x", fontsize=15)
plt.ylabel("f(x)", fontsize=15)
plt.plot(x_1, f(x_1), color='blue', linewidth=5)

# 2 Derivative 
plt.subplot(1,2,2)
plt.xlim(-2, 3)
plt.ylim(-3,8)
plt.title("Slope of the Cost Function", fontsize=17)
plt.xlabel("x", fontsize=15)
plt.ylabel("d(x)", fontsize=15)
plt.grid()
plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)

plt.show()
```
