## Linear Regression using python

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('lsd-math-score-data.csv')
```

```python
data
```

```python
time = data[['Time_Delay_in_Minutes']]
LSD = data[['LSD_ppm']]
score = data[['Avg_Math_Test_Score']]
```

```python
  plt.title("Tissue concentration of LSD over time", fontsize=15)

  plt.xlabel("Time in minutes")
  plt.ylabel("Tissue LSD ppm")
  plt.text(x=0.5, y=-0.5, s="Wagner et al. (1968)")

plt.plot(time, LSD)
plt.show()
```

```python
regr = LinearRegression()

regr.fit(LSD, score)

print('Theta1 :', regr.coef_[0][0])
print('Intercept: ', regr.intercept_[0])

# To find out how much of the variation in our data is explained by the amount of drugs in the volunteers tissue
# we call the score() method on our data
print('R-squrared:', regr.score(LSD, score))

# To predict our score
predicted_score = regr.predict(LSD)

```

```python
# Challenge: add title 'Arithmetic vs LSD-25'
# Add label on x Axis 'Tissue LSD ppm'
# Add label on Y Axis 'Performance Score'
plt.title('Arithmetic vs LSD-25', fontsize=15)
plt.xlabel('Tissue LSD ppm')
plt.ylabel('Performance Score')
plt.xlim(1, 7)
plt.ylim(25, 85)
plt.style.use('fivethirtyeight')

plt.scatter(LSD, score, alpha=0.6, s=100, c='green')
plt.plot(LSD, predicted_score)
plt.show()
```
