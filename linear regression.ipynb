```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost-revenue-clean.csv')
```

```python
data.describe()
```

```python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.title('Movie Cost Vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

## Linear Regression

```python
regression = LinearRegression()
regression.fit(X, y)
```

Slope Coefficient 

```python
regression.coef_
```

```python
# intercept
regression.intercept_
type(regression.intercept_)
```

```python
plt.figure(figsize=(10,6))           # set the size of the graph
plt.scatter(X, y, alpha=0.3)
plt.plot(X, regression.predict(X), color="red", linewidth=2)   # Calculate a prediction for each of our budget data and plot it on the graph. 

plt.title('Movie Cost Vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
regression.score(X, y)     # Goodness of fit also called R squared
```
